```python
import pandas as pd
import seaborn as sns
computers = pd.read_csv("Computers.csv")
```

```python
sns.histplot(computers["price"])
```

```python
sns.histplot(computers["speed"])
```

```python
sns.histplot(computers["ram"])
```

```python
import matplotlib.pyplot as plt

columns = ["price", "speed", "ram"]
data = computers

fig, ax = plt.subplots(len(columns), layout ='tight')

for col, ax in zip(columns, ax.flat):
  sns.histplot(data=data, x=col, ax=ax)
  ax.set_title(f"Histogram of {col}")
```

```python
sns.heatmap(computers.corr(numeric_only=True), annot=True, vmin = -1, vmax= 1, cmap="coolwarm")
```

```python
sns.pairplot(computers.select_dtypes("number"), corner=True)
## using corner to remove the mirror image and to get 1 chart per relationship
```

```python
sns.pairplot(computers.select_dtypes("number").sample(500), corner=True)
## To sped up, we have taken sample from dataset and removed the mirror image
```

```python
sns.lmplot(computers, x='ram', y="price")
## +ve slope so higher ram higher priced.
```

```python
sns.lmplot(computers, x="ram", y="price", hue="premium")
```

```python
sns.lmplot(computers.query("3<ram<11"), x="ram", y="price", hue="premium")
##when ram is greater than 3 and less than 11, we can consider both slopes almost equal and not much different so can't say both the slopes are different
```
