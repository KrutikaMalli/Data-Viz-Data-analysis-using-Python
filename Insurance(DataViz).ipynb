```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insurance = pd.read_csv("insurance.csv")
```

```python
insurance.head()
```

```python
insurance.info()
```

```python
insurance.describe()
```

```python
insurance["smoker"].value_counts()
```

```python
insurance["smoker"].value_counts().plot.bar()
```

```python
sns.heatmap(insurance.corr(numeric_only=True), annot=True, vmin=1, vmax=-1, cmap="coolwarm")
```

```python
sns.scatterplot(insurance, x="age", y="charges", hue="smoker")
```

```python
insurance.groupby("smoker")["charges"].mean().plot.bar()
```

```python
sns.pairplot(insurance)
```

```python
sns.pairplot(insurance, corner=True)
# Hides the mirror images and gives us clearer image
```

```python
sns.pairplot(insurance, corner=True, diag_kind = "kde")
# To get smoother curve distributions, so instead of histograms...we have kde curves
```

```python
sns.lmplot(insurance, x="bmi", y="charges")
## We can see huge chuck or set of data points not passing through the line
```

```python
sns.lmplot(insurance, x="bmi", y="charges", hue="smoker")
##graph shows that bmi has a very strong impact on charges if you are a smoker.
## slope of line is different for someone who is a smoker versus non smoker vs bmi
```
