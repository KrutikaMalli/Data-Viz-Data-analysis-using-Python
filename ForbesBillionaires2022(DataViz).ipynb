#*Forbes Billionaires Analysis using Python*#

The dataset used in this is about the billionaires around the world which was curated by Forbes and can be downloaded from kaggle.

It contains the following information about the billionaires:

1) Name

2) Networth

3) Country

4) Source

5) Rank

6) Age

7) Industry

Importing necessary python libraries and dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("ForbesBillionaires2022.csv")
print(data.head())
```

Checking to see if the dataset contains any missing values

```python
print(data.isnull().sum())
```

There are no missing values in this dataset.

Now, if we look at the Networth column, we have to convert this column to float and strip away the "$" sign at the starting and "B" at the ending.

```python
data["Networth"]=data["Networth"].str.strip("$")
data["Networth"]=data["Networth"].str.strip("B")
data["Networth"]=data["Networth"].astype(float)
```

```python
print(data.head())
```

Let's look at the top 10 billionares based on their networth:


```python
df = data.sort_values(by = ["Networth"], ascending=False).head(10)
plt.figure(figsize=(25,7))
sns.histplot(x="Name", hue="Networth", data=df)
plt.title("Top 10 Billionaires 2022" , fontsize =30)
plt.show()
```

Let's look at the top 5 domains having max number of billionaires:

```python
print(data["Source"].value_counts().head())
```

```python
a = data["Source"].value_counts().head()
index = a.index
sources = a.values
custom_colors = ["skyblue", "yellowgreen", "tomato", "blue", "yellow"]
plt.figure(figsize=(5,5))
plt.pie(sources, labels=index, colors=custom_colors)
central_circle = plt.Circle((0,0), 0.5, color="white")
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=10)
plt.title("Top 5 Domains to become a  Billionaire", fontsize=12)
plt.show()

```

Now, let's look at top 5 industries with the most number of billionaires:

```python
print(data["Industry"].value_counts().head())
```

```python
a = data["Industry"].value_counts().head()
index = a.index
industries = a.values
custom_colors = ["Purple", "yellow", "yellowgreen", "red", "pink"]
plt.figure(figsize=(5,5))
plt.pie(industries, labels=index, colors = custom_colors)
central_circle = plt.Circle((0,0),0.5, color="white")
fig=plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size =8)
plt.title("Top 5 Industries with most number of Billionaires", fontsize=12)
plt.show()
```

Now, let's look at top 5 countries with most number of billionaires:

```python
print(data["Country"].value_counts().head())
```

```python
a = data["Country"].value_counts().head()
index = a.index
Countries = a.values
custom_colors = ["Red", "Blue", "Silver", "Pink", "yellow"]
plt.figure(figsize=(5,5))
plt.pie(Countries, labels=index, colors=custom_colors)
central_circle=plt.Circle((0, 0), 0.5, color="White")
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size = 12)
plt.title("Top 5 countries with most billionaires", fontsize=20)
plt.show()
```
